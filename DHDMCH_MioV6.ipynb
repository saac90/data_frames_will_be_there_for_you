# **1. Introduzione**

Questo progetto analizza un dataset di opere d’arte, con l’obiettivo di esplorare alcuni pattern storici relativi alla produzione artistica e alla rappresentazione all’interno della collezione.

Ma già dalla prima occhiata mi son resso conto ch'erano tutti uomini... E quindi mi son chiesto se c'erano delle donne.

```python
df["artisti"].str.contains("femmina", na=False).any()
```

### **3.1.2. Procesamento dei Dati**

E infatti c'erano, ma soltanto 36. E allora perché? **Sara che negli primi anni non c'erano artiste ma poi si sono livellati?** E quindi ho iniziato a cercare sugli anni d'attività di queste artiste in un'altra colonna dal DataFrame (data_creazione) che ho dovuto trasformare in numeri.



```python
df_donne = df[df["artisti"].str.contains("femmina", na=False)].copy()
```

```python
df_donne["anno"] = df_donne["data_creazione"].str.extract(r'(\d{4})')
df_donne["anno"] = df_donne["anno"].astype(float)
```

```python
attivita_artiste = (
    df_donne
    .groupby("artisti")["anno"]
    .agg(["min", "max", "count"])
    .sort_values("min")
)

attivita_artiste
```

Qui ho pasato da *float* a *int*.

```python
attivita_clean = attivita_artiste.dropna(subset=["min", "max"]).copy()
```

```python
attivita_clean[["min", "max"]] = attivita_clean[["min", "max"]].astype(int)
```

```python
attivita_clean
```

E qui ho dropato il (femmina).

```python
attivita_clean.index = attivita_clean.index.str.replace(" (femmina)", "", regex=False)
```

```python
attivita_clean
```

### **3.1.3. Explanatory Data Analysis**

Bene. Adesso vediamo di aggruppargli per secoli, così vediamo si aumentano come abbiamo presuposto prima.

Per fare questo ho creato un'altra colonna pulita solo con la data piu antica di ognuna, da dove prendo il secolo.

```python
attivita_clean["siglo"] = ((attivita_clean["min"] - 1) // 100 + 1).astype(int)

attivita_clean[["min", "siglo"]]
```

Non sembra che ci sia una cresciutta sostenuta nel tempo, pero proviamo di vizzualizarla in un grafico (ChatGPT mi ha consigliato di usare matplotlib).

```python
import matplotlib.pyplot as plt

artistas_por_siglo = attivita_clean.groupby("siglo").size()

plt.figure(figsize=(8,5))
artistas_por_siglo.plot(marker="o", linestyle="-")
plt.xticks(artistas_por_siglo.index)
plt.ylim(0, artistas_por_siglo.max() + 1)
plt.xlabel("Secoli")
plt.ylabel("Numero di artiste")
plt.title("Artiste donne rappresentate nella collezione per secolo")
plt.grid(True)
plt.show()
```

È strano che la quantità di donne non aumenti costantemente con i secoli; **e vero che non ci sono piu artiste nel XX secolo o e un bias dalla collezione?** Proviamo a vedere se è correlata con quella degli uomini (per questo ho creato una copia dal *df* originale in modo di non modificarlo).

```python
df_artisti = df[["artisti", "data_creazione"]].copy()

df_artisti["genere"] = df_artisti["artisti"].str.extract(r"\((maschio|femmina)\)")
df_artisti["artisti_limpio"] = df_artisti["artisti"].str.replace(r"\s*\(maschio\)|\s*\(femmina\)", "", regex=True)
```

```python
df_artisti["anno"] = df_artisti["data_creazione"].str.extract(r"(\d{4})").astype(float)
```

```python
df_artisti = df_artisti.dropna(subset=["anno", "genere"]).copy()
df_artisti["siglo"] = ((df_artisti["anno"] - 1) // 100 + 1).astype(int)
```

Ecco gli artisti donne e uomini nei secoli:

```python
artisti_por_siglo_genere = df_artisti.groupby(["siglo", "genere"]).size().unstack(fill_value=0)
artisti_por_siglo_genere
```

```python
artisti_por_siglo_genere.plot(marker="o", linestyle="-", figsize=(8,5))
plt.xlabel("Secolo")
plt.ylabel("Numero di artisti")
plt.title("Confronto di rappresentatività per secolo: donne contro uomini")
plt.grid(True)
plt.show()
```

 Risulta evidente che ci siano piu uomini, ma quello che mi interessa e controllare se le proporzione (non i numeri) vanno assieme.

```python
totales = df_artisti["genere"].value_counts()

artisti_por_siglo_genere = df_artisti.groupby(["siglo", "genere"]).size().unstack(fill_value=0)

prop_genere = artisti_por_siglo_genere.copy()
prop_genere["femmina"] = prop_genere["femmina"] / totales["femmina"]
prop_genere["maschio"] = prop_genere["maschio"] / totales["maschio"]

prop_genere
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(prop_genere.index, prop_genere["femmina"], marker="o", linestyle="-", color="purple", label="Donne")

plt.plot(prop_genere.index, prop_genere["maschio"], marker="o", linestyle="-", color="blue", label="Uomini")

plt.xticks(range(prop_genere.index.min(), prop_genere.index.max()+1))
plt.ylim(0, max(prop_genere.max())*1.1)
plt.xlabel("Secolo")
plt.ylabel("Proporzione sul totale del genere")
plt.title("Tendenza storica della rappresentatività per genere nella collezione")
plt.legend()
plt.grid(True)

plt.show()
```

### **3.1.4. Conclusione**

Anche se il numero assoluto di uomini è molto più alto, quando guardo la proporzione di ciascun genere sul totale, **i *pattern* storici risultano piu o meno paralleli**: di solito i secoli con meno uomini hanno anche meno donne, e così via.

⚥ Questo indica che la collezione riflette, per quanto possibile, l’attività storica di ciascun genere: non sembra che la collezione abbia volutamente ignorato le donne nei secoli in cui erano più presenti, né che le abbia esagerate dove erano poche.

♂   Quello che emerge chiaramente è che ci sono pochissime donne tra XI e XV secolo, e questo riflette la realtà storica della produzione artistica e dell’accesso al mondo dell’arte.

♀   Nel XIX secolo noto un aumento relativo della rappresentazione femminile in proporzione al totale, coerente con il fatto che più donne iniziano a essere documentate.

⚥   Nel XXI secolo non c'è nessuna donna nel DataFrame, ma ci sono anche pochi uomini; cioè la collezione non ha ancora incorporato abbastanza opere contemporanee.

**Quindi non noto un *bias* aggiuntivo evidente nella collezione rispetto al genere**: la proporzione di rappresentazione segue più o meno l’attività storica documentata per ciascun genere. **Tuttavia, la collezione riflette il bias storico generale: la maggior parte degli artisti documentati e rappresentati continuano ad essere uomini**.

## **3.2. *The one about religion***

Cosi come c'e una sottorapresentanza femminile nell'arte, ci sara anche una sovrarapresentanza dalla religione cattolica nell'arte italiano? E si sì, come cambiato questo nel tempo?

### **3.2.1. Procesamento dei Dati**

Ho deciso di creare una nuova colonna chiamata "tematica" con solo due valori: "religiosa" e "laica". Così son partito dalla colonna esistente "titoli" per ottenere i risultati attraverso diversi *keywords*.

```python
keywords_religiose = [
    "religione", "fede", "fedele", "fedeli",
    "dio",
    "caino", "abele", "noè", "golia", "oloferne", "isacco",
    "cristo", "christus", "gesù",
    "miracolo", "miracoli", "moltiplicazione dei pani",
    "croce", "crocifisso", "crocefissione", "crocifissione", "crocifissino",
    "maria", "madonna", "vergine", "immacolata",
    "visitazione", "annunciazione", "natività",
    "battista", "battesimo",
    "santo", "santa", "san", "sant", "santi",
    "angelo", "angeli", "arcangelo", "arcangeli",
    "maddalena",
    "pietà", "ultima cena", "adorazione", "preghiera",
    "deposizione", "assunzione", "resurrezione",
    "evangelista", "evangelisti",
    "apostolo",
    "chiesa", "papa", "cardinale", "cardinali", "vescovo", "vescovi", "vescovile", "diacono",
    "savonarola",
    "sacra", "sacro", "paolo",
    "prodigo",
    "fuga in egitto",
    "visione", "martirio", "martire", "misericordia",
]
```

```python
print(keywords_religiose)
```

```python
len(keywords_religiose)
```

```python
sorted(keywords_religiose)
```

Ho creato la nuova colonna.

```python
import numpy as np

pattern = '|'.join(keywords_religiose)

df['tematica'] = np.where(
    df['titolo'].str.lower().str.contains(pattern, na=False),
    'religiosa',
    'laica'
)
```

E controllato attraverso *sample* se c'erano opere ancora non riconosciute come religiose (ho fatto e rifatto quest'operazione varie volte fino a non trovare nessuna).

```python
df[df['tematica'] == 'laica'][['titolo']].sample(50)
```

Il risultato è stato il seguente:

```python
df['tematica'].value_counts()
```

```python
df['tematica'].value_counts()

df['tematica'].value_counts(normalize=True) * 100
```

### **3.2.2. Exploratory Data Analysis**

Cioè una maggioranza d'opere di carattere religioso ma non troppo definita. Allora ho pensato si questo equilibrio era costante oppure cambiava con i tempi. E quindi sono ritornato alla colonna "data_creazione" per scoprirlo.

```python
df['data_creazione_str'] = df['data_creazione'].astype(str)
```

```python
df['anno'] = df['data_creazione_str'].str.extract(r'(\d{4})')
df['anno'] = pd.to_numeric(df['anno'], errors='coerce')
```

```python
df['secolo'] = (df['anno'] - 1) // 100 + 1
```

```python
df[['data_creazione', 'anno', 'secolo']].head(10)
```

### **3.2.3. Explanatory Data Analysis**

Adesso si c'e qualcosa d'interesante: i percentuali d'opere religiose/laiche cambia nel tempo.

```python
df_validi = df.dropna(subset=['anno'])
tabella_prop = pd.crosstab(df_validi['secolo'], df_validi['tematica'], normalize='index') * 100
tabella_prop
```

```python
import matplotlib.pyplot as plt

# Convertir índice a entero
tabella_prop.index = tabella_prop.index.astype(int)

# Ordenar por siglo
tabella_prop = tabella_prop.sort_index()

# Graficar
tabella_prop.plot(kind='bar', stacked=True, figsize=(10,6), color=['#66b3ff','#ff9999'])

plt.ylabel("Percentuale (%)")
plt.xlabel("Secolo")
plt.title("Distribuzione opere religiose e laiche per secolo")
plt.legend(title="Tematica")
plt.xticks(rotation=0)
plt.ylim(0,100)  # asegurarse de que las barras estén entre 0 y 100%
plt.show()
```

### **3.2.4. Conclusione**

✝️ XI secolo: tutte le opere registrate sono religiose, riflettendo l'**assoluta dominanza dell'arte ecclesiastica** nell'Alto Medioevo.

✝️ Secolo XIII-XVII: **predominanza di opere religiose**, anche se appare già una certa percentuale di opere laiche; questo riflette il ruolo centrale della Chiesa e delle commissioni religiose durante il Basso Medioevo e il Rinascimento.

🖼️ XVIII-XIX secolo: **predominanza di opere laiche**, indicando un cambiamento verso il mercato privato, l'incarico di collezionisti privati e l'interesse per temi profani nell'Illuminismo e nel Romanticismo.

XXI secolo: i dati sono molto limitati; sebbene la proporzione di opere religiose sia ancora elevata, non è possibile trarre conclusioni significative a causa della bassa rappresentatività del campione.
